**Data Extracation  **

```python
# Transaction Data of User from different states
agg_trans = "C:/Users/Workstation/Documents/Phonepe/pulse/data/aggregated/transaction/country/india/state/"
agg_states_list = os.listdir(agg_trans)

col = { 'State' : [], 'Year' : [], 'Quater' : [], 'Transaction_type' : [], 'Transaction_count' : [], 'Transaction_amount' : []}
# "col" easy way to navigate in pandas
for i in agg_states_list:
    trans_i = agg_trans + i +"/"    #"/" ---> file path in windows
    agg_year = os.listdir(trans_i)  # os.listdir ---> gives all the files in an specifed diectory.

    for j in agg_year:
        trans_j = trans_i + j + "/"
        agg_year_lst = os.listdir(trans_j)

        for k in agg_year_lst:
            trans_k = trans_j + k 
            #print(trans_k)

        # converting data to Tables using Pandas
            Data = open(trans_k, 'r') # 'r' ---> Read mode
            transaction = json.load(Data)
            for l in transaction ['data'] ['transactionData']:
                Name = l['name']
                count = l['paymentInstruments'][0]['count']
                amount = l['paymentInstruments'][0]['amount']
                col['Transaction_type'].append(Name)
                col['Transaction_count'].append(count)
                col['Transaction_amount'].append(amount)
                col['State'].append(i)
                col['Year'].append(j)
                col['Quater'].append(int(k.strip('.json')))
aggregated_transaction = pd.DataFrame(col)

print(aggregated_transaction)

'''
aggregated_transaction.to_csv('aggregated_transaction.csv')
aggregated_transaction.to_csv('C:/Users/Workstation/Documents/Phonepe/data extracted/Agg_Trans.csv')
'''

```

```python
Reg_users = "C:/Users/Workstation/Documents/Phonepe/pulse/data/map/user/hover/country/india/state/"
map_list = os.listdir(Reg_users)

col = {"State": [], "Year": [], "Quater": [], "District": [],
        "RegisteredUser": []}
for i in map_list:
    user_i = Reg_users + i + "/"
    Agg_yr = os.listdir(user_i)

    for j in Agg_yr:
        user_j = user_i + j + "/"
        Agg_yr_list = os.listdir(user_j)

        for k in Agg_yr_list:
            user_k = user_j + k
            #print(user_k)
            Data = open(user_k, 'r')
            D = json.load(Data)

            for user in D["data"]["hoverData"].items():
                district = user[0]
                registereduser = user[1]["registeredUsers"]
                col["District"].append(district)
                col["RegisteredUser"].append(registereduser)
                col['State'].append(i)
                col['Year'].append(j)
                col['Quater'].append(int(k.strip('.json')))
map_user = pd.DataFrame(col)
map_user
#map_user.to_csv('C:/Users/Workstation/Documents/Phonepe/data extracted/district_registering_map.csv')
```
